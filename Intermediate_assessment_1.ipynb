**1.** Create a DataFrame `df` from this dictionary `data` which has the index `labels`.

````python
import numpy as np
import pandas as pd


data = {
    'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
    'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
    'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']
}
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

# Create DataFrame

df=pd.DataFrame(data,index=labels)
df
````

````python
import matplotlib.pyplot as plt
import pandas as pd

# Filter rows where delays are not null
filtered_data = data.dropna(subset=['dep_delay', 'arr_delay'])

# Scatter plot for dep_delay vs arr_delay
plt.figure(figsize=(10, 6))
plt.scatter(filtered_data['dep_delay'], filtered_data['arr_delay'], alpha=0.5, s=10, color='blue')
plt.title('Scatter Plot of Departure Delay vs Arrival Delay')
plt.xlabel('Departure Delay (minutes)')
plt.ylabel('Arrival Delay (minutes)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
````
