```python
from sklearn.impute import SimpleImputer

# Separate numerical and categorical columns
num_cols = titanic.select_dtypes(include=['float64', 'int64']).columns
cat_cols = titanic.select_dtypes(include=['object']).columns
```

```python
# Fill missing values: Median for numerical, Most Frequent for categorical
num_imputer = SimpleImputer(strategy='median')
cat_imputer = SimpleImputer(strategy='most_frequent')
```

```python
titanic[num_cols] = num_imputer.fit_transform(titanic[num_cols])
titanic[cat_cols] = cat_imputer.fit_transform(titanic[cat_cols])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Function to plot boxplots before and after handling outliers
def plot_outliers(data, column):
    # Create subplots
    fig, ax = plt.subplots(1, 2, figsize=(12, 6))

    # Plot before handling outliers
    sns.boxplot(data=data, y=column, ax=ax[0])
    ax[0].set_title(f'Before Handling Outliers: {column}')

    # Handle outliers
    handle_outliers(data, column)

    # Plot after handling outliers
    sns.boxplot(data=data, y=column, ax=ax[1])
    ax[1].set_title(f'After Handling Outliers: {column}')

    plt.tight_layout()
    plt.show()

# Plot boxplots for 'Age', 'Fare', 'SibSp'
for column in ['Age', 'Fare', 'SibSp']:
    plot_outliers(titanic, column)
```
